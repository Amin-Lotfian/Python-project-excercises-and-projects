#  DROPPP

```python
df['job_country'].dropna(axis=0,inplace=True)
```

```python
df.dropna(subset=['job_country'], inplace=True)
```

```python
df_Ireland = df[df['job_country']=='Ireland']
```

```python
df_Ireland['job_posted_month']=df_Ireland['job_posted_date'].dt.strftime('%B')
df_Ireland
```

```python
df_Ireland.pivot_table(
    index='job_posted_month',
    columns='job_title_short',
    aggfunc='size'
)
```

```python
df_Ireland.groupby('job_posted_month')['job_location'].count()
```

```python
df_Ireland.info()
```

```python
df_job_country_salary = df_Ireland.pivot_table(
    index='job_posted_month',
    columns='job_title_short',
    aggfunc='size'
)

job_titles = ['Data Analyst', 'Data Scientist', 'Data Engineer']
df_job_country_salary = df_job_country_salary[job_titles]


month_order = ['January', 'February', 'March', 'April', 'May', 'June',
              'July', 'August', 'September', 'October', 'November', 'December']
df_job_country_salary = df_job_country_salary.reindex(month_order)
df_job_country_salary


df_job_country_salary.plot(kind='line')
plt.ylabel('job count')
plt.xlabel('')
plt.title('monthly job postings for top data jobs in Ireland')
plt.xticks(rotation=45, ha='right')
plt.show()
```
