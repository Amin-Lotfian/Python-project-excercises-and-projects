```python
# Importing Libraries
!pip install datasets
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
df= df.dropna(subset=['salary_year_avg'])

df_exploded = df.explode('job_skills')

df_exploded = df_exploded.groupby('job_skills')['salary_year_avg'].agg(['median', 'count']).sort_values(by='count', ascending=False)

df_exploded = df_exploded.rename(columns={'count':'skill_count' ,'median':'median_salary'})

df_exploded['skill_percent'] = (df_exploded['skill_count'] / (df['job_title_short'].count())) * 100

df_exploded = df_exploded [df_exploded['skill_percent'] > 5]

df_exploded.sort_values(by='skill_percent', ascending=False)
```

```python
df_exploded.plot(kind='scatter', x='skill_percent', y='median_salary')

# plt.text() function to label the points
for i, txt in enumerate(df_exploded.index):
    plt.text(df_exploded['skill_percent'].iloc[i], df_exploded['median_salary'].iloc[i], txt)
plt.text(35,100000, 'example')
plt.xlabel('Count of Job Postings')
plt.ylabel('Median Yearly Salary')
plt.title('Salary vs. Count of Top Skills in Data Analyst Jobs')



from matplotlib.ticker import PercentFormatter
ax=plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
ax.xaxis.set_major_formatter(PercentFormatter(decimals=0))

plt.tight_layout()
plt.show()
```

# combine all dictionaries into one

```python
df_technology = df['job_type_skills'].copy()

# remove duplicates
df_technology = df_technology.drop_duplicates()

# remove NaN values
df_technology = df_technology.dropna()

# combine all dictionaries into one
technology_dict = {}
for row in df_technology:
    row_dict = ast.literal_eval(row)  # convert string to dictionary
    for key, value in row_dict.items():
        if key in technology_dict:  # if key already exists in technology_dict, add value to existing value
            technology_dict[key] += value
        else:                       # if key does not exist in technology_dict, add key and value
            technology_dict[key] = value

# remove duplicates by converting values to set then back to list
for key, value in technology_dict.items():
    technology_dict[key] = list(set(value))

technology_dict
```

```python
df_technology2 = pd.DataFrame(list(technology_dict.items()), columns=['Technology', 'skills'])
df_technology2 = df_technology2.explode('skills')
df_technology2
```
