# *Good excercise*

everywhere you put iloc, from where the new filters will be applied:

```python
df2[['job_title','salary_year_avg']].iloc[90:12000][(df2.job_title_short == 'Data Analyst') & (df2.salary_year_avg>100000) & (df2.salary_year_avg.notna())]
```

```python
df3= df2[['job_title','salary_year_avg']].iloc[90:12000][(df2.job_title_short == 'Data Analyst') & (df2.salary_year_avg>100000) & (df2.salary_year_avg.notna())]
mean_salary = df3['salary_year_avg'].mean()
std_salary = df3['salary_year_avg'].std()
sum_salary = df3['salary_year_avg'].sum()

print(f"Mean: {mean_salary}")
print(f"Standard Deviation: {std_salary}")
print(f"Sum: {sum_salary}")
```

```python
df2[['job_title','salary_year_avg']][(df2.job_title_short == 'Data Analyst') & (df2.salary_year_avg>100000) & (df2.salary_year_avg.notna())].iloc[90:12000]
```
