```python
# Install datasets Library (if not already installed)
!pip install datasets

# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  # we will use this for plotting

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

```

```python
median_salary_year = df['salary_year_avg'].median()
median_salary_hour = df['salary_hour_avg'].median()
```

```python
df_filled = df
```

```python
id(df_filled) == id(df)
```

```python
df_filled = df.copy()
```

```python
df_filled ['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_year)
df_filled ['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_salary_hour)
```

```python
df_unique = df_filled.drop_duplicates()
df_unique
```

```python
id(df_unique) == id(df_filled)
```

```python
df_unique = df_unique.drop_duplicates(subset=['job_title', 'company_name', 'job_location'])
df_unique
```
